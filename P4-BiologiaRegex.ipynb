##### 2.- Crea dos archivos. Uno almacena los accession que empiezen por 'a' y el otro el resto

```python
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']
```

```python
a_accs = [acc for acc in accs if acc.startswith('a')]
resto_accs = [acc for acc in accs if not acc.startswith('a')]

with open('accs_a.txt', 'w') as f_a:
    for acc in a_accs:
        f_a.write(acc + '\n')

with open('accs_resto.txt', 'w') as f_r:
    for acc in resto_accs:
        f_r.write(acc + '\n')

print("Archivos creados correctamente:")
print("- accs_a.txt contiene:", a_accs)
print("- accs_resto.txt contiene:", resto_accs)
```

```python
def porcentaje_A(secuencia):
    if len(secuencia) == 0:
        return 0
    else:
        secuencia = secuencia.upper()
        total = len(secuencia)
        conteo_A = secuencia.count('A')
        porcentaje = (conteo_A / total) * 100
        return porcentaje

assert porcentaje_A("") == 0

print("La funcion OG es correcta")
```
