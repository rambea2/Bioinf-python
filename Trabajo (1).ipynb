#### 5.- Devuelve la secuencia de ARN que surgiría a partir de la siguiente secuencia. Recuerda que la secuencia de ARN resultante tiene que ser complementaria a la de ADN. (1 pto)

```python
# primero creamos un diccionario que mapea cada base de ADN a su complementaria en ARN
# Las keys son las bases de ADN y los values son las bases complementarias en ARN
transcripcion = {
    'A': 'U',  # adenina → uracilo
    'T': 'A',  # timina → adenina
    'C': 'G',  # citosina → guanina
    'G': 'C'   # guanina → citosina
}

# vamos a recorrer cada base de la secuencia de ADN
secuencia_ARN = ''  # empezamos con la cadena de ARN vacía
for base in secuencia:
    # buscamos en el diccionario la base complementaria
    complementaria = transcripcion[base]
    # le agregamos la base complementaria a la secuencia de ARN
    secuencia_ARN += complementaria

# imprimo solo los primeros 1000 nucleótidos del ARN porque sino la pagina me deja de funcionar
print("Secuencia de ARN (primeros 1000 nucleótidos):")
print(secuencia_ARN[:1000])

# también muestro la longitud total
print("\nLongitud total de la secuencia de ARN:", len(secuencia_ARN))
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
# Codones de parada
stop_codons = ["TAA", "TAG", "TGA"]
# Aminoácidos aromáticos
aromaticos = ["F", "Y", "W"]

def complemento(dna):
    """Devuelve la strand negativa (complementaria e invertida)"""
    tabla = str.maketrans("ATCG", "TAGC")
    return dna.translate(tabla)[::-1]

def encontrar_orfs(secuencia):
    """
    Encuentra ORFs en las 6 posibles frames y calcula la mejor proteína.
    Devuelve:
    - Todos los ORFs
    - Top 6 ORFs más largos
    - Mejor predicción (más largo)
    """
    secuencia = secuencia.upper()
    dna_rev = complemento(secuencia)
    
    orfs = []

    def extraer_orfs(seq, strand):
        for frame in range(3):
            codons = [seq[i:i+3] for i in range(frame, len(seq)-2, 3)]
            inicio = None
            for i, codon in enumerate(codons):
                if codon == "ATG" and inicio is None:
                    inicio = i
                if codon in stop_codons and inicio is not None:
                    orf_seq = ''.join(codons[inicio:i+1])
                    orfs.append({
                        "strand": strand,
                        "frame": frame+1,
                        "start": frame + inicio*3,
                        "end": frame + i*3 + 3,
                        "dna": orf_seq
                    })
                    inicio = None
            # ORF hasta el final si no hay stop
            if inicio is not None:
                orf_seq = ''.join(codons[inicio:])
                orfs.append({
                    "strand": strand,
                    "frame": frame+1,
                    "start": frame + inicio*3,
                    "end": len(seq),
                    "dna": orf_seq
                })

    # ORFs de la strand positiva
    extraer_orfs(secuencia, "+")
    # ORFs de la strand negativa
    extraer_orfs(dna_rev, "-")

    # Ordenamos por longitud
    orfs = sorted(orfs, key=lambda x: len(x["dna"]), reverse=True)

    # Top 6 ORFs más largos
    top6 = orfs[:6]

    # Mejor predicción = el más largo
    mejor = orfs[0] if orfs else None

    # Calcular % de aminoácidos aromáticos (F, Y, W)
    if mejor:
        prot = []  # traducir DNA a proteína simplificado (codones)
        codon_tabla = {
            # Codones simplificados para aromáticos y stop
            "TTT":"F","TTC":"F","TTA":"L","TTG":"L",
            "CTT":"L","CTC":"L","CTA":"L","CTG":"L",
            "ATT":"I","ATC":"I","ATA":"I","ATG":"M",
            "GTT":"V","GTC":"V","GTA":"V","GTG":"V",
            "TAT":"Y","TAC":"Y","TAA":"*","TAG":"*",
            "TGA":"*","TCT":"S","TCC":"S","TCA":"S",
            "TCG":"S","CCT":"P","CCC":"P","CCA":"P",
            "CCG":"P","ACT":"T","ACC":"T","ACA":"T",
            "ACG":"T","GCT":"A","GCC":"A","GCA":"A",
            "GCG":"A","TGT":"C","TGC":"C","TGG":"W",
            "CGT":"R","CGC":"R","CGA":"R","CGG":"R",
            "AGT":"S","AGC":"S","AGA":"R","AGG":"R",
            "GGT":"G","GGC":"G","GGA":"G","GGG":"G",
            "CAT":"H","CAC":"H","CAA":"Q","CAG":"Q",
            "AAT":"N","AAC":"N","AAA":"K","AAG":"K",
            "GAT":"D","GAC":"D","GAA":"E","GAG":"E"
        }
        for i in range(0, len(mejor["dna"])-2, 3):
            codon = mejor["dna"][i:i+3]
            aa = codon_tabla.get(codon, "X")  # X = desconocido
            if aa == "*":
                break
            prot.append(aa)
        prot = ''.join(prot)
        arom_count = sum(prot.count(a) for a in aromaticos)
        arom_pct = arom_count / len(prot) * 100 if len(prot) > 0 else 0
        rel_len = len(prot)*3 / len(secuencia) * 100
        mejor["proteina"] = prot
        mejor["aromatic_pct"] = arom_pct
        mejor["rel_len_pct"] = rel_len

    return orfs, top6, mejor

# Demostracion del uso

secuencia_orf = "ATGAAATGAAAATGTAGATGTTTTAAATG"  # mi secuencia
todos_orfs, top6_orfs, mejor_orf = encontrar_orfs(secuencia_orf)

print("Top 6 ORFs más largos")
for i, o in enumerate(top6_orfs,1):
    print(f"ORF {i}: Strand {o['strand']}, Frame {o['frame']}, Longitud: {len(o['dna'])}")

if mejor_orf:
    print("\n Mejor predicción ")
    print(f"Strand: {mejor_orf['strand']}, Frame: {mejor_orf['frame']}")
    print(f"Longitud ORF: {len(mejor_orf['dna'])}")
    print(f"Proteína: {mejor_orf['proteina']}")
    print(f"% Aromáticos: {mejor_orf['aromatic_pct']:.2f}")
    print(f"Longitud relativa a ADN: {mejor_orf['rel_len_pct']:.2f}%")
```
